# Plot diagrams


```python
import matplotlib.pyplot as plt
```

```python
celestial_objects = [
    {"name": "Jupiter", "mass": 1.898e27, "radius": 69911},
    {"name": "Saturn", "mass": 5.683e26, "radius": 58232},
    {"name": "Uranus", "mass": 8.681E+25, "radius": 25362},
    {"name": "Neptun", "mass": 1.024E+26, "radius": 24622},
    {"name": "Sirius B", "mass": 2.025E30, "radius": 5.5E6},
    {"name": "Sun", "mass": 1.989E30, "radius": 696340},
]

for celestial_object in celestial_objects:
    plt.scatter(label=celestial_object["name"], x=celestial_object["mass"], y=celestial_object["radius"])

plt.legend()
plt.gca().set_xlabel("Mass (kg)")
plt.gca().set_ylabel("Radius (km)")
plt.show()
```
